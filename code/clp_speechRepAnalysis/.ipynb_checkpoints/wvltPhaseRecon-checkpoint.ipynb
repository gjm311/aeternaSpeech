```python
# create plot area or use the one provided
def showScalogram(time, coefs=None, freqs=None, hop=50, spectrum='amp', ax=None, cscale='linear', cmap='jet', clim=None,
                  cbar='vertical', cbarlabel=None,cbarkw=None,figsize=(10, 4.0),yaxis='frequency', xlim=None, ylim=None, yscale='log', coi=False,
                  ylabel="Frequency", xlabel="Time (s)", title=None):
    
    CBAR_DEFAULTS = {'vertical'   : { 'aspect':30, 'pad':0.03, 'fraction':0.05 },'horizontal' : { 'aspect':40, 'pad':0.12, 'fraction':0.05 }}

    COI_DEFAULTS = {'alpha':'0.5','hatch':'/'}
    
    FS=16000
    
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)

    # adjust y axis ticks
    scales_period = np.divide(1,freqs)  # needed also for COI mask
    xmesh = np.concatenate([time, [time[-1]+hop]])
    if yaxis == 'period':
        ymesh = np.concatenate([scales_period, [scales_period[-1]]])
        ylim  = ymesh[[-1,0]] if ylim is None else ylim
        ax.set_ylabel("Period" if ylabel is None else ylabel)
    elif yaxis == 'frequency':
        df    = freqs[-1]/freqs[-2]
        ymesh = np.concatenate([freqs, [freqs[-1]*df]])
        # set a useful yscale default: the scale freqs appears evenly in logscale
        yscale = 'log' if yscale is None else yscale
        ylim   = ymesh[[-1, 0]] if ylim is None else ylim
        ax.set_ylabel("Frequency" if ylabel is None else ylabel)
        #ax.invert_yaxis()
    else:
        raise ValueError("yaxis must be one of 'frequency' or 'period', found "
                          + str(yaxis)+" instead")

    # limit of visual range
    xr = [time.min(), time.max()]
    if xlim is None:
        xlim = xr
    else:
        ax.set_xlim(*xlim)
    if ylim is not None:
        ax.set_ylim(*ylim)
        
    # adjust logarithmic scales on request (set automatically in Frequency mode)
    if yscale is not None:
        ax.set_yscale(yscale)

    # choose the correct spectrum display function and name
    if spectrum == 'amp':
        values = np.abs(coefs)
        sp_title = "Amplitude"
        cbarlabel= "abs(CWT)" if cbarlabel is None else cbarlabel
    elif spectrum == 'real':
        values = np.real(coefs)
        sp_title = "Real"
        cbarlabel= "real(CWT)" if cbarlabel is None else cbarlabel
    elif spectrum == 'imag':
        values = np.imag(coefs)
        sp_title = "Imaginary"
        cbarlabel= "imaginary(CWT)" if cbarlabel is None else cbarlabel
    elif spectrum == 'power':
        sp_title = "Power"
        cbarlabel= "abs(CWT)$^2$" if cbarlabel is None else cbarlabel
        values = np.power(np.abs(coefs),2)
    elif hasattr(spectrum, '__call__'):
        sp_title = "Custom"
        values = spectrum(coefs)
    else:
        raise ValueError("The spectrum parameter must be one of 'amp', 'real', 'imag',"+
                         "'power' or a lambda() expression")

    # labels and titles
    ax.set_title("Continuous Wavelet Transform "+sp_title+" Spectrum"
                 if title is None else title)
    ax.set_xlabel("Time/spatial domain" if xlabel is None else xlabel )
    if max(time)/FS>1:
        ax.set_xticklabels(map(str, np.linspace(0,max(time)/FS,6, dtype=np.int)))
    else:
        ax.set_xticklabels(map(str, np.linspace(0,(max(time)/FS)*1000,10, dtype=np.int)))
        ax.set_xlabel('Time (ms)')

    if cscale == 'log':
        isvalid = (values > 0)
        cnorm = LogNorm(values[isvalid].min(), values[isvalid].max())
    elif cscale == 'linear':
        cnorm = None
    else:
        raise ValueError("Color bar cscale should be 'linear' or 'log', got:"+
                         str(cscale))

    # plot the 2D spectrum using a pcolormesh to specify the correct Y axis
    # location at each scale
    qmesh = ax.pcolormesh(xmesh, ymesh, values, cmap=cmap, norm=cnorm)

    if clim:
        qmesh.set_clim(*clim)

    # fill visually the Cone Of Influence
    # (locations subject to invalid coefficients near the borders of data)
    if coi:
        # convert the wavelet scales frequency into time domain periodicity
        scales_coi = scales_period
        max_coi  = scales_coi[-1]

        # produce the line and the curve delimiting the COI masked area
        mid = int(len(xmesh)/2)
        time0 = np.abs(xmesh[0:mid+1]-xmesh[0])
        ymask = np.zeros(len(xmesh), dtype=np.float16)
        ymhalf= ymask[0:mid+1]  # compute the left part of the mask
        ws    = np.argsort(scales_period) # ensure np.interp() works
        minscale, maxscale = sorted(ax.get_ylim())
        if yaxis == 'period':
            ymhalf[:] = np.interp(time0,
                  scales_period[ws], scales_coi[ws])
            yborder = np.zeros(len(xmesh)) + maxscale
            ymhalf[time0 > max_coi]   = maxscale
        elif yaxis == 'frequency':
            ymhalf[:] = np.interp(time0,
                  scales_period[ws], 1./scales_coi[ws])
            yborder = np.zeros(len(xmesh)) + minscale
            ymhalf[time0 > max_coi]   = minscale
        elif yaxis == 'scale':
            ymhalf[:] = np.interp(time0, scales_coi, scales)
            yborder = np.zeros(len(xmesh)) + maxscale
            ymhalf[time0 > max_coi]   = maxscale
        else:
            raise ValueError("yaxis="+str(yaxis))

        # complete the right part of the mask by symmetry
        ymask[-mid:] = ymhalf[0:mid][::-1]

        # plot the mask and forward user parameters
        plt.plot(xmesh, ymask)
        coikw = COI_DEFAULTS if coikw is None else coikw
        ax.fill_between(xmesh, yborder, ymask, **coikw )

    # color bar stuff
    if cbar:
        cbarkw   = CBAR_DEFAULTS[cbar] if cbarkw is None else cbarkw
        colorbar = plt.colorbar(qmesh, orientation=cbar, ax=ax, **cbarkw)
        if cbarlabel:
            colorbar.set_label(cbarlabel)

    return ax
```

```python
def create_wavelets(num_samples,nbf=8,dil=2):
    """
    Compute wavelet basis function for 'morl' wavelets
    ADAPTED from Phase retrieval for wavelet transforms: 
    http://www-math.mit.edu/~waldspur//wavelets_phase_retrieval.html#source_code

    :param num_samples: number of samples (should be equivalent to signal that will be transformed)
    :param dil: dilation factor between wavelets.
    :returns freqs: frequencies related to wavelets.
    :returns psi: 3D family of wavelets array with each wavelet being a row.
    :returns phi: gaussian low freq. function.
    """

    old_settings = np.seterr(under='ignore', over='ignore')

#     nbf=int(np.ceil(np.log(dil*num_samples/2)/np.log(dil)))
    freqs=[(num_samples/2)*np.power(1/dil,bf) for bf in np.arange(nbf)]

    ms=np.tile(freqs,(num_samples,1)).T
    oms=np.tile(np.arange(num_samples),(nbf,1))

    p=.3/(np.log(dil)/np.log(2))**2
    c=np.exp(-p) 

    d=1
    for k in range(10):
        d=1/(1-np.exp(-p*2*d))
       
    exp1=-p*(np.divide(d*oms,(ms-1))**2)
    exp2=-p*(np.divide(d*oms,ms)**2)

    psi=np.exp(exp1)-c*np.exp(exp2)

    max_psi=np.max(abs(psi))
    psi=np.float32(np.divide(psi,max_psi))
    phi=np.exp(-np.arange(num_samples)**2)

    return freqs,psi,phi
```

```python
def wavelet_transform(signal,psi,phi=None,dil=2):
    """
    Compute wavelet transform of a signal
    ADAPTED from Phase retrieval for wavelet transforms: 
    http://www-math.mit.edu/~waldspur//wavelets_phase_retrieval.html#source_code
    
    :param signal: signal to be transformed
    :param psi: family of wavelets with each wavelet being a row.
    :param phi: gaussian low freq. function.
    :returns fpsi: components of the wavelet transform (time domain rep.).
    :returns fphi: conv. of signal with low-freq. filter
    """
    fphi=phi
    
    
#     if len(signal)<psi.shape[1]:
#         psi=create_wavelets(len(signal),psi.shape[0],dil)
        
#     signal=signal.T
    
    if len(signal) != np.shape(psi)[1]:
        print('The sizes of the wavelets and the signal do not match!')
        
    ft_sig=np.fft.fft(signal.T)
    fpsi=np.fft.ifft(np.multiply(psi,ft_sig))
    fphi=np.fft.ifft(np.multiply(phi,ft_sig))
    
    return fpsi,fphi
    

    
```

```python
def dual_wvlts(psi):
    """
    This function computes the dual of a wavelet family
    
    
    :param psi: wavelets
    :returns psi_conj: dual of wavelets
    """
    
    
    nbf,num_samples=psi.shape
    
    sum_squares=np.sum(abs(psi)**2,axis=1)
    sum_squares=np.where(sum_squares<10e-14,sum_squares,1)
    
    return np.divide(np.conj(psi),np.tile(sum_squares,(nbf,1)))
```

```python
def exp_decay_approx(psi,dil):
    """
    Determines whether the wavelets decay exponentially or not;
    if they do not, compute the indexes where they will be truncated
    

    :param psi: wavelets
    :param dil: dilation factor
    
    :returns exp_decay: dictionary  with boolean if decay is exponential, and 
    if not, truncation index.
    """
    
    nbf=psi.shape[0]
    num_samples=psi.shape[1]
    
    log_psi=-np.log(np.divide(abs(psi),5+1e-13))
    ref=np.asarray([dil*ii for ii in range(nbf)]).T
    ref=np.multiply(np.dot(ref,np.arange(num_samples)),2/(num_samples-1))
    
    min_log=np.min(np.divide(log_psi[ref>0 & abs(psi)>1e-12],ref[ref>0 & abs(psi)>1e-12]))
    
    if min_log<np.log(10)/5:
        exp_decay={'bool':False,'cutoff':0}
        
        for s in range(nbf):
            is_small=(abs(psi[s,:])<.02*np.max(abs(psi[s,:])))
            stays_small=np.fliplr(np.cumprod(np.fliplr(is_small)))
            
            if max(stays_small<1):
                cutoff[s]=num_samples
            else:
                forg=max(stays_small)
                idx=np.where(stays_small==max(stays_small))
                cutoff[s]=idx
                
        exp_decay['cutoff']=cutoff
            
    else:
        exp_decay={'bool':True}
        
    
    
```

```python
def gerch_sax(fpsi,psi,max_iters,rod):
    """
    This function attempts to reconstruct a function from the modulus
    of its wavelet transform by the method of Gerchberg and Saxton
    (alternate projections, described in 'A practical algorithm for the
    determination of phase from image and diffraction plane pictures]' (Optik, 1972))
    
    
    :param fpsi: modulus of the wavelet transform
    :param psi: wavelets
    :param max_iters: max number of iterations
    :param rod: rate of display of progress
    
    :returns f_rec: reconstructed function
    """
    
    num_samples=fpsi.shape[1]
    fpsi=abs(fpsi)
    
    n_ff=np.floor(num_samples/2)+1
    n_sub=min(num_samples,3*n_ff)
    
    ff_rec=np.random.normal(0,1,n_sub)+1j*np.random.normal(0,1,n_sub)
    ff_rec[0]=0
    ff_rec[n_ff:]=0
    f_rec=np.fft.ifft(ff_rec)
    
    psi_conj=dual_wvlts(psi)
    psi_conj[:,n_ff:]=0
    
    for k in range(max_iters):
        if k%rod==0:
            print("iteration "+str(k)+" of Gerchberg-Saxton algorithm...")
            
        #Project the wavelet transform of f_rec on the set of functions 
        #whose moduli correspond to the desired ones
        psi_rec = np.fft.ifft(np.multiply(psi,ff_rec,axis=1))
        psi_rec = np.divide(np.multiply(fpsi,psi_rec),(abs(psi_rec)+1e-13))

        #Project psi_rec on the set of admissible wavelet transforms
        psi_rec_f = np.fft.fft(psi_rec)
        ff_rec = np.sum(np.multiply(psi_rec_f,psi_conj),axis=1)
        
        
    return np.fft.ifft(ff_rec,num_samples)
                              
```

```python
def find_alpha(psi1,psi2,psi3,dil,exp_decay,j0):
    """
    :param psi1: wavelet of maximal freq.
    :param psi2: wvlt of freq under psi1
    :param psi3: wvlt of freq under psi2
    :param dil: dilation factor
    :param j0: index of psi1 (psi1=psi[j0,:])
    
    :returns alpha: real number such that psi1 .* exp(-[0:N-1] * alpha)
      appears to be optimal as an intermediate wavelet in the 
      reconstruction process.
    :returns upper_alpha: max reasonable value for alpha
    """
    
    num_samples=psi1.shape[1]
    
    psi1_sq=np.fft.fft(abs(np.fft.ifft(psi1))**2)
    
    #maximal value
    supp_psi1=[ps for ps in psi1_sq[:int(num_samples//2)] if ps > .05*np.max(abs(psi1_sq))]
    size_supp=max(supp_psi1)
    upper_alpha=1/max(1,size_supp)*np.log(50)
    
    nb_it=5   #number of iters
    prec=6   #number of values tested at each iter
    
    alpha_min=0
    alpha_max=upper_alpha
    oms=np.arange(N)
    if exp_decay['bool']:
        oms2=oms
    else:
        oms2=exp_decay['cutoff'][j0]-abs(exp_decay['cutoff'][j0]-oms)
    
    
    for jj in range(nb_it):
        alphas=alpha_min+np.divide(np.arange(prec),prec-1)*(alpha_max-alpha_min)
        
        for ii in range(prec):
            psi1_m=abs(np.multiply(psi1,np.exp(np.dot(-oms,alphas(ii)))))   #low wvlt. candidate
            psi2_m=abs(np.multiply(psi2,np.exp(np.dot(oms2,alphas(ii))*dil)))   #high wvlt. candidate
            psi3_m=abs(np.multiply(psi3,np.exp(np.dot(oms,alphas(ii))*dil**2)))   ##low freq. function

            psi1_m=psi1_m/np.linalg.norm(psi1_m)
            psi2_m=psi2_m/np.linalg.norm(psi2_m)
            psi3_m=psi3_m-np.dot(sum(np.multiply(psi2_m,psi3_m)),psi2_m)  
            psi3_m=psi3_m/np.linalg.norm(psi3_m)

            norm_diff=1-sum(np.multiply(psi2_m,psi1_m))**2 -sum(np.multiply(psi3_m,psi2_m))**2

        forg=np.min(norm_diff)
        i_min=np.where(forg==np.min(norm_diff))

        i_inf=max(1,i_min-1)
        i_sup=min(prec,i_min+1)

        alpha_min=alphas[i_inf]
        alpha_max=alphas[i_sup]
        
    return alphas[i_min],upper_alpha
```

```python
def compute_high_low(psi,alpha,dil,exp_decay):
    """
    Coputes high, low psi
    ADAPTED from Phase retrieval for wavelet transforms: 
    http://www-math.mit.edu/~waldspur//wavelets_phase_retrieval.html#source_code
    
    
    :param psi: family of wavelets with each wavelet being a row.
    :param alpha: exponential scaling factor
    :param dil: dilation factor
    """
    
    nbf=psi.shape[0]
    num_samples=psi.shape[1]
    
    oms_low=[dil**rn for rn in np.arange(nbf)]
    if exp_decay['bool']:
        oms_high=oms_low
    else:
        oms_high=np.zeros((nbf,num_samples))
        for s in range(nbf):
            ind=exp_decay['cutoff'][s]
            oms_high[s,:]=[dil**s*(ind-abs(itr-ind)) for itr in range(num_samples)]
    
    oms_low=alpha*oms_low
    oms_high=alpha*oms_high
    forg=np.max(abs(psi),axis=1)
    ind=[np.where(f==rowM) for f,rowM in zip(forg,psi)]
    ind=[ii[0][0] for ii in ind]
    ind=np.ravel_multi_index((np.arange(nbf).T,ind), psi.shape)
    
    vals_high=oms_high[ind]
    oms_high=oms_high-vals_high
    oms_low=oms_low-vals_high
    
    psi_low=np.multiply(np.exp(-np.min(oms_low,20)))
    psi_high=np.multiply(np.exp(np.min(oms_high,20)))
    
    return psi_low,psi_high
```

```python
def eval_prod():
    """
    Compute true value of opt_psi_low*conj(opt_psi_high) from squared 
    modulus of opt_psi.
    ADAPTED from Phase retrieval for wavelet transforms: 
    http://www-math.mit.edu/~waldspur//wavelets_phase_retrieval.html#source_code
    
    
    :param sq_mod: squared noisy modulus of opt psi.
    :param al: reral number for optimization of psi.
    :param ind: regularization parameter
    
    :returns product: product of opt psis (low,high)
    """
    
    sq_mod=np.asarray(sq_mod).T
    num_samples=sq_mod.shape[1]
    oms=np.concatenate((np.arange((num_samples//2)+1), np.arange(-num_samples//2,0)))
    
    ff_sq_mod=np.fft.fft(sq_mod)
    
    oms_trunc=ind+abs(oms-ind)
    product=fft.fft.ifft(np.multiply(ff_sq_mod,np.exp(-np.multiply(al*oms_trunc))))
    
    return product  
    
```

```python
def eval_prod_ref(sq_mod,al,true_prod):
    """
    Compute the value of the correct regularization parameter to use
    in eval_prod, by testing several values on an example where the
    ground true product is known and keeping the best value.
    ADAPTED from Phase retrieval for wavelet transforms: 
    http://www-math.mit.edu/~waldspur//wavelets_phase_retrieval.html#source_code
    
    
    :param sq_mod: squared noisy modulus of opt psi.
    :param al: reral number for optimization of psi.
    :param true_prod: true value of opt_psi_low*conj(opt_psi_high)
    
    :returns ind: best regularization parameter
    :returns err_min: minimum error induced by ind
    :returns product: product of opt psis (low,high)
    """
    
    num_samples=len(sq_mod)
    i_inf=-num_samples/2
    ind=-num_samples/4
    prec=4
    
    err=np.zeros((1,prec+1))
    
    while i_sup-i_inf>-i_sup/50:
        
        i_test=i_inf+np.multiply(np.arange(prec+1),(i_sup-i_inf)/prec)
        
        for s in range(prec+1):
            product=eval_prod(sq_mod,al,i_test[s])
            err[s]=np.linalg.norm(true_prod-product)
            
        #returning value for which error is minimal
        err_min=min(err)
        i_min=np.where(err_min==min(err))
        i_inf=i_test[np.max(1,i_min-1)]
        i_sup=i_test[np.min(prec+1,i_min+1)]
        ind=i_test[i_min]
        
    return ind,err_min,product
    
```

```python
def reconstruct(fpsi,psi,freqs,dil):
    """
    Determines whether the wavelets decay exponentially or not.
    If they do not, compute the indexes where they will be truncated.
    ADAPTED from Phase retrieval for wavelet transforms: 
    http://www-math.mit.edu/~waldspur//wavelets_phase_retrieval.html#source_code
    
    
    :param psi: family of wavelets with each wavelet being a row.
    :param dil: dilation factor between wavelets.
    """
    
    nbf = 16
    num_samples = psi.shape[1]
    max_iters=10000
    gs_rough_iters=50   #number of Gerchberg-Saxton [G.S.] iters allowed for fisrt approximation of signal
    gs_iters=2000   #number of G.S. iters allwoed for final refinement.
    rod=1000   #rate of display
    rod_gs=50   #rate of display during G.S.
    tol=0.999
    noise_norm=1e-3   #relative norm of the noise.
    
    exp_decay=exp_decay_approx(psi,dil)
    
    f_ref=gerch_sax(fpsi,psi,max_iters,rod)
    f_ref_f=np.fft.fft(f_ref)
    f_ref_f[0]=0
    f_ref_f[num_samples+1:]=0
    f_ref=np.fft.ifft(f_ref_f)
    
    fpsi_ref=wavelet_transform(f_ref,psi)
    
    #Add gaussian white noise
    noise_psi=np.random.normal(0,1,len(fpsi_ref))
    noise_norm=np.sqrt(np.sum(abs(noise_psi))**2)
    transf_norm=np.sqrt(np.sum(abs(fpsi_ref))**2)
    fpsi_noisy_ref=abs(fpsi_ref+noise_norm*transf_norm/noise_norm*noise_psi)
    
    #calculate reference error
    err_ref=np.sqrt(np.sum(abs(abs(fpsi_noisy_ref)**2 - abs(fpsi_ref)**2)**2 ,axis=1))
    err_ref=np.divide(err_ref,np.sqrt(np.sum(abs(fpsi_ref)**2,axis=1)))
    inds_ref=np.where(err_ref<1e-3)
    
    #choose alpha based off middle-freq. wvlt (w/index j0)
    #helps avoid truncation or aliasing for high/low frequency wvlts rspt.
    j0=np.min(nbf-2,np.round(np.sqrt(nbf)))
    alpha,alpha_max=find_alpha(psi[j0-1,:],psi[j0,:],psi[j0+1,:],dil,exp_decay,j0)
    alpha=alpha*(1/dil**(j0-1))
    alpha_max=alpha_max*(1/dil**(j0-1))
    
    alphas=[alpha*2**(.1*ex) for ex in np.arange(11)]
    alphas=[alph for alph in alphas if alph<alpha_max]
    
    if inds_ref.any() and len(alphas)>1 and min(inds_ref)<nbf-np.round(np.log(2)/np.log(dil)):
        print("Choosing optimal alpha...")
        
        true_prod_ref=np.zeros((len(alphas),num_samples))
        err_tmp=np.zeros((1,len(alphas)))
        err_alpha=np.zeros((1,len(alphas)))
        
        for n_alpha in range(len(alphas)):
            
            #we will select the highest value of alpha for which the
            #error in the presence of noise is not too high
            psi_low,psi_high=compute_high_low(psi,alphas(n_alpha),dil,exp_decay)

            #renormilization
            cs=np.sqrt(np.divide(np.max(abs(psi_high),axis=1),np.max(abs(psi_low),axis=1)))
            psi_low=np.multiply(psi_low,cs)
            psi_high=np.multiply(psi_high,np.divide(1,cs))
            
            #wvlt transform of f_ref
            fpsi_ref=wavelet_transform(f_ref,psi)
            fpsi_ref_low=wavelet_transform(f_ref,psi_low)
            fpsi_ref_high=wavelet_transform(f_ref,psi_high)
            #add noise
            noise_psi=np.random.normal(0,1,len(fpsi_ref))
            noise_norm=np.sqrt(np.sum(abs(noise_psi))**2)
            transf_norm=np.sqrt(np.sum(abs(fpsi_ref))**2)
            fpsi_noisy_ref=abs(fpsi_ref+noise_norm*transf_norm/noise_norm*noise_psi)
            
            #computing error on products
            for s in inds_ref:
                true_prod_ref[s,:]=np.multiply(fpsi_ref_low[s,:],np.conj(fpsi_ref_high[s,:]))
                forg,err_prod = eval_prod_ref(abs(fpsi_noisy_ref[s,:])**2,alphas[n_alpha]*dil**s,true_prod_ref[s,:])
                err_tmp[s]=err_prod/np.linalg.norm(true_prod_ref[s,:])
            
            err_alpha[n_alpha]=max(err_tmp)
            
        
        
```
