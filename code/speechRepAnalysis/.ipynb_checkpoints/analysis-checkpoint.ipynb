```python
from AEspeech import AEspeech
import os
from scipy.io.wavfile import read
import numpy as np
```

```python
wav_path = "../tedx_spanish_corpus/speech/"

wav_file=wav_path+os.listdir(wav_path)[0]
fs, signal=read(wav_file)

aespeech=AEspeech("CAE", 1024, fs) # load the pretrained CAE with 1024 units
mat_spec=aespeech.compute_spectrograms(wav_file) # compute the decoded spectrograms from the autoencoder
print(mat_spec.size())
aespeech.show_spectrograms(mat_spec)
```
